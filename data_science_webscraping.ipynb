__Tworzenie zapytań do API__

```python
response = requests.get(
    'https://api.github.com/search/repositories',
    params={
        'q': 'requests+language:python'
    }
)
```

```python
json_response = response.json() # Odpowiedź json przerobiona na słownik pythonowy
```

```python
repositories = json_response['items']
```

```python
import pandas as pd
```

```python
pd.DataFrame(repositories)
```
