```python
url = 'https://www.allsides.com/media-bias/media-bias-ratings'
r=requests.get(url)
```

```python
soup=BeautifulSoup(r.content, 'html.parser')
```

```python
views_table=soup.find(class_='views-table')
```

```python
views_table
```

```python
rows = views_table.select('tbody tr')
```

```python
rows
```

```python
name=rows[0].find(class_='source-title')
```

```python
name.get_text().strip()
```

```python
href = rows[0].select_one('.source-title a')
```

```python
href
```

```python
names = [row.find(class_='source-title').get_text().strip()
        for row in rows]
```

```python
base_url = 'https://www.allsides.com'
links = [f"{base_url}{row.select_one('.source-title a')['href']}"
         for row in rows]
```

```python
links
```

```python
biases = [(row
           .select_one('.views-field-field-bias-image a')['href']
           .split('/')[-1])
          for row in rows
         ]
```

```python
biases
```

```python
agree = rows[0].select_one('.agree').get_text().strip()
```

```python
int(agree)
```

```python
agree_ratios = []
for row in rows:
    agree = row.select_one('.agree').text
    disagree = row.select_one('.disagree').text
    agree_ratio = int(agree) / int(disagree)
    agree_ratios.append(agree_ratio)
```

```python
agree_ratios
```

```python
def get_agreeance_text(ratio):
    if ratio > 3: return "absolutely agrees"
    elif 2 < ratio <= 3: return "strongly agrees"
    elif 1.5 < ratio <= 2: return "agrees"
    elif 1 < ratio <= 1.5: return "somewhat agrees"
    elif ratio == 1: return "neutral"
    elif 0.67 < ratio < 1: return "somewhat disagrees"
    elif 0.5 < ratio <= 0.67: return "disagrees"
    elif 0.33 < ratio <= 0.5: return "strongly disagrees"
    elif ratio <= 0.33: return "absolutely disagrees"
    else: return None
    
print(get_agreeance_text(2.5))
```

```python
agreeance_texts = [get_agreeance_text(ratio)
                  for ratio in agree_ratios]
```

```python
agreeance_texts
```

```python
pd.DataFrame({
    'name': names,
    'allsides_page': links,
    'bias': biases,
    'agree_ratio': agree_ratios,
    'agreeance_text': agreeance_texts
})
```
