```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births.head()
```

```python
births.describe()
```

```python
births.dtypes
```

```python
"""select m.year, m, f from
(select year, sum(births) as m from births where gender = "m" group by year) as m
join (select year, sum(births) as f from births where gender = "f" group by year) as f
on m.year = f.year
""";
```

```python
births.pivot_table('births', index='year', columns='gender',
                   aggfunc='sum').plot()

plt.ylabel('total births per year')
```

```python
births['decade'] = 10 * (births['year'] // 10)

births.pivot_table('births', index='decade', columns='gender',
                   aggfunc='sum').plot()
plt.ylabel('total births per decade')
```

```python
quartiles = np.percentile(births['births'], [25, 50, 75])
mu = quartiles[1]
sig = 0.74 * (quartiles[2] - quartiles[0])
```

```python
births = births.dropna().copy()
births['day'] = births['day'].astype(int)
```

```python
births = births.query('(births > @mu - 5*@sig) & (births < @mu + 5*@sig)')
```

```python
births.describe()
```

```python
births.head()
```

```python
births.index = pd.to_datetime(10000 * births.year 
               + 100 * births.month 
               + births.day,
               format='%Y%m%d')
```

```python
births['dayofweek'] = births.index.dayofweek
```

```python
births.pivot_table('births', index='dayofweek',
                   columns='decade', aggfunc='mean').plot()
plt.gca().xaxis.set_major_locator(
    mpl.ticker.FixedLocator(range(7))
)
plt.gca().set_xticklabels(['Mon', 'Tues',
                           'Wed', 'Thurs',
                           'Fri', 'Sat', 'Sun'])
plt.ylabel('mean births by day of week');
```

```python
births_by_date = births.pivot_table('births',
                  [births.index.month,
                  births.index.day])
```

```python
from datetime import datetime
```

```python
births_by_date.index = [datetime(2012, month, day)
                       for (month, day) in births_by_date.index]
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)

style = dict(size=10, color='gray')
ax.text('2012-1-1', 3950, "New Year's day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas ", ha='right', **style)

ax.set(title='USA births by day of year (1969-1988)',
       ylabel='avg daily births');

ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'))
```

```python
[int(x) for x in '01:05:38'.split(':')]
```

```python
from datetime import timedelta

def convert_time(s):
    h, m, s = [int(x) for x in s.split(':')]
    return timedelta(hours=h, minutes=m, seconds=s)
```

```python
convert_time("01:05:38")
```

```python
marathon_df = pd.read_csv(marathon_url, converters={
    'split': convert_time,
    'final': convert_time
})
```

```python
marathon_df.dtypes
```

```python
marathon_df['split_sec'] = marathon_df['split'].dt.seconds
marathon_df['final_sec'] = marathon_df['final'].dt.seconds
```

```python
marathon_df.head()
```

```python
with sns.axes_style('white'):
    g = sns.jointplot(x=marathon_df['split_sec'],
                  y=marathon_df['final_sec'],
                  kind='hex')
    g.ax_joint.plot(np.linspace(4000, 16000),
                   np.linspace(8000, 32000), ':k')
```

```python
g = sns.PairGrid(marathon_df, vars=['age', 'split_sec', 'final_sec'],
                hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
```

```python
sns.violinplot(x=marathon_df['gender'], y=marathon_df['final_sec'],
              palette=['lightblue', 'lightpink']);
```

```python
marathon_df['age_dec']= 10 * (marathon_df['age'] // 10)
```

```python
sns.violinplot(x=marathon_df['age_dec'],
               y=marathon_df['final_sec'],
               hue='gender',
               data=marathon_df,
               split=True,
               inner='quartille',
               palette=['lightblue', 'lightpink']
              );
```
