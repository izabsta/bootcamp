```python
# zadanie9 
"""
SELECT * 
FROM flights LEFT JOIN planes 
    ON flights.tailnum=planes.tailnum
"""
```

```python
planes_df = dfs['planes']
flights_df = dfs['flights']
pd.merge(flights_df, planes_df, on = 'tailnum', how = 'left')
```

```python
# Zadanie 10
"""
SELECT planes.*, airlines.* FROM
(SELECT DISTINCT carrier, tailnum FROM flights) AS cartail
        INNER JOIN planes ON cartail.tailnum=planes.tailnum
        INNER JOIN airlines ON cartail.carrier=airlines.carrier
"""
```

```python
planes_df = dfs['planes']
airlines_df = dfs['airlines']
flights_df = dfs['flights']

cartail_df = flights_df[['carrier', 'tailnum']].drop_duplicates()

cartail_planes_df = pd.merge(cartail_df, planes_df, on = 'tailnum')

cartail_planes_airlines_df = pd.merge(cartail_planes_df, airlines_df, on = 'carrier')

cartail_planes_airlines_df
```
